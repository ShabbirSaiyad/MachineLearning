**Aim: Implement “AND” gate using Perceptron Learning (Self Implementation)** 

```python
import numpy as np

# Dataset (AND gate)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([0, 0, 0, 1]) 

# Initialize parameters
w = np.zeros(2)
b = 0
lr = 0.1  

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Predicted class
def predict_value(z):
    return 1 if sigmoid(z) >= 0.5 else 0

# Training
for _ in range(5000):  
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        a = sigmoid(z)
        y_pred = predict_value(z)

        # Gradient of sigmoid loss
        error = y[i] - a
        w = w + lr * error * X[i]
        b = b + lr * error

print("Final Weights:", w)
print("Final Bias:", b)

print("\nTesting AND with Sigmoid Activation:")
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    print(f"{X[i]} -> {predict_value(z)}")
```
