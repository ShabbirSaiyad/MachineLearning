```python
import numpy as np
```

**i) Simple Linear Regression using normal equation**

```python
# Reproducible seed
np.random.seed(42)

# Generate synthetic data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add bias term (intercept column)
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape (100,2)

# Normal Equation: theta = (X^T X)^(-1) X^T y
theta_normal = np.linalg.inv(X_b.T @ X_b) @ (X_b.T @ y)

# Compute R^2 score
y_hat = X_b @ theta_normal
mse = np.mean((y - y_hat) ** 2)

print("Normal Equation coefficients:", theta_normal.flatten())

print("MSE:", mse)
```

**Multiple Linear Regression using normal equation**

```python
# Reproducible results
np.random.seed(42)

# Generate data with 3 features
X = 2 * np.random.rand(100, 3)
y = 4 + 3*X[:, 0] + 3*X[:, 1] + 3*X[:, 2] + np.random.randn(100)

# Add bias (intercept term)
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape (100, 4)

# Normal Equation
theta = np.linalg.inv(X_b.T @ X_b) @ (X_b.T @ y)

# Predictions
y_pred = X_b @ theta

mse = np.mean((y - y_pred) ** 2)

print("Coefficients (intercept + 3 features):", theta)
print("MSE:", mse)
```

**ii) Simple and Multiple Linear Regression using Gradient Descent method without using sklearn.**

```python
print("Choose which Linear Regression you want to go for:")
print("1. Simple Linear Regression")
print("2. Multiple Linear Regression")

def switch_example(value):
    if value == 1:
        # Simple Linear Regression: y = 4 + 3X + noise
        X = 2 * np.random.rand(100, 1)
        true_theta_simple = np.array([4, 3]).reshape(-1, 1)
        y = 4 + 3 * X + np.random.randn(100, 1)
        return (X, y, true_theta_simple)

    elif value == 2:
        # Multiple Linear Regression: y = 4 + 3x1 + 3x2 + 3x3 + noise
        X = 2 * np.random.rand(100, 3)
        true_theta_multiple = np.array([4, 3, 3, 3]).reshape(-1, 1)
        y = 4 + 3*X[:, 0] + 3*X[:, 1] + 3*X[:, 2] + np.random.randn(100)
        y = y.reshape(-1, 1)
        return (X, y, true_theta_multiple)

    else:
        return "Invalid choice. Please select 1 or 2."

# Get user input
choice = int(input("Enter 1 or 2: "))

result = switch_example(choice)

if isinstance(result, tuple):
    X, y, true_theta = result
    print("\nSelected Option:", choice)
    print("X shape:", X.shape)
    print("y shape:", y.shape)
    print("True Theta:\n", true_theta)
else:
    print(result)
```

```python
#Numbers of rows
m = len(y)
print(m)
```

```python
# --- Split into train (80) and test (20) ---
split_idx = int(0.8 * m)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
```

```python
# Add bias term
X_train_b = np.c_[np.ones((len(X_train), 1)), X_train] 

X_test_b = np.c_[np.ones((len(X_test), 1)), X_test]    
```

```python
class GDRegressor:

    def __init__(self, learning_rate=0.1, epochs=100):
        self.learning_rate = learning_rate   
        self.epochs = epochs
        self.theta = None

    def fit(self, X_train, y_train):
        # Initialize theta
        self.theta = np.random.randn(X_train.shape[1], 1)
        
        for i in range(self.epochs):
            y_pred_train = X_train.dot(self.theta)
            g = (2 / len(y_train)) * X_train.T.dot(y_pred_train - y_train)
            self.theta = self.theta - self.learning_rate * g
            
        print("\nPredicted Theta:\n", self.theta)
        
    def predict(self, X_test):
        X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        return X_b.dot(self.theta)
         

    def mse(self, y_true, y_pred):
        #Compute Mean Squared Error
        return np.mean((y_true - y_pred) ** 2)
```

```python
gd = GDRegressor(learning_rate = 0.1, epochs = 1290);
```

```python
gd.fit(X_train_b,y_train)
```

```python
y_pred = gd.predict(X_test)

mse_value = gd.mse(y_test, y_pred)
print("\nMSE:", mse_value)
```

```python
class SGDRegressor:

    def __init__(self, learning_rate = 0.1, epochs = 100):
        self.learning_rate = learning_rate   
        self.epochs = epochs
        self.theta = None

    def fit(self, X_train, y_train):
        # Initialize theta
        self.theta = np.random.randn(X_train.shape[1], 1)
        
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                
                idx = np.random.randint(0, X_train.shape[0])
                
                X_train_1_row = X_train[ idx : idx + 1]
                y_train_1_row = y_train[ idx : idx + 1]

                y_pred_train = X_train_1_row.dot(self.theta)
                
                g = 2 * X_train_1_row.T.dot( y_pred_train - y_train_1_row )
                
                self.theta = self.theta - self.learning_rate * g
        print("\nPredicted Theta:\n", self.theta)
        
    def predict(self, X_test):
        X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        return X_b.dot(self.theta)
        

    def mse(self, y_true, y_pred):
        #Compute Mean Squared Error
        return np.mean((y_true - y_pred) ** 2)
```

```python
class MBGDRegressor:

    def __init__(self, learning_rate=0.1, epochs=100, batch_size=32):
        self.learning_rate = learning_rate   
        self.epochs = epochs
        self.batch_size = batch_size
        self.theta = None

    def fit(self, X_train, y_train):
        n_samples, n_features = X_train.shape
        self.theta = np.random.randn(X_train.shape[1], 1)

        for i in range(self.epochs):
            # Shuffle data
            indices = np.arange(X_train.shape[0])
            np.random.shuffle(indices)
            X_train = X_train[indices]
            y_train = y_train[indices]

            # Process in mini-batches
            for start_idx in range(0, n_samples, self.batch_size):
                end_idx = start_idx + self.batch_size
                X_batch = X_train[start_idx:end_idx]
                y_batch = y_train[start_idx:end_idx]

                y_pred = X_batch.dot(self.theta)

                # Gradient of MSE loss
                g = (2 / len(y_batch)) * X_batch.T.dot(y_pred - y_batch)

                # Update theta
                self.theta = self.theta - self.learning_rate * g

        print("\nPredicted Theta:\n", self.theta)

    def predict(self, X_test):
        X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        return X_b.dot(self.theta)
        
    def mse(self, y_true, y_pred):
        #Compute Mean Squared Error
        return np.mean((y_true - y_pred) ** 2)

   
```

```python
mbgd = MBGDRegressor(learning_rate=0.01, epochs=51, batch_size=int(X_train_b.shape[0]/10))
```

```python
mbgd.fit(X_train_b,y_train)
```

```python
y_pred = mbgd.predict(X_test)

mse_value = mbgd.mse(y_test, y_pred)
print("\nMSE:", mse_value)
```
