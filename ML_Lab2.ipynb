```python
url_titanic = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url_titanic);
titanic
```

```python
# Replacing values in dataframe
# Replace values in Pclass
d1 = titanic["Pclass"].replace({1: "First", 2: "Second", 3: "Third"});
print("\nReplaced Pclass values:");
print(d1.head());

# Replace values in Sex
d2 = titanic["Sex"].replace(["female", "male"], ["Woman", "Man"]);
print("\nReplaced Sex values:");
print(d2.head());
```

```python
#inplace data replacement
# Replace directly in dataset
titanic["Sex"].replace(["female", "male"], ["Woman", "Man"], inplace=True);

print("\nTitanic dataset after inplace replacement in 'Sex':");
print(titanic[["Name", "Sex"]].head());
```

```python
#Renaming columns

#temporary name
print("Temporary");
d3 = titanic.rename(columns={"Sex":"Gender"});
print(d3[["Name", "Gender"]].head());

# Inplace rename
print("\nInplace");
titanic.rename(columns={"Sex": "Gender"}, inplace=True);
print(titanic[["Name", "Gender"]].head());

# Multiple columns rename
print("\nMultiple columns:");
titanic.rename(columns={"Ticket": "PassengerTicket", "Age": "PassengerAge"}, inplace=True);
print(titanic[["PassengerTicket","PassengerAge"]].head(2));
```

```python
# finding Min,Max,Count,Avg, and Sum

print("Mean Age:", titanic["PassengerAge"].mean());
print("Max Age:", titanic["PassengerAge"].max());
print("Min Age:", titanic["PassengerAge"].min());
print("Sum Age:", titanic["PassengerAge"].sum());
print("Count Age:", titanic["PassengerAge"].count());

print("\nVariance:", titanic["PassengerAge"].var());
print("Standard Deviation:", titanic["PassengerAge"].std());
print("Kurtosis:", titanic["PassengerAge"].kurt());
print("Skewness:", titanic["PassengerAge"].skew());
print("SEM:", titanic["PassengerAge"].sem());
print("Mode:\n", titanic["PassengerAge"].mode().head());
print("Median:", titanic["PassengerAge"].median());
```

```python
#  Unique values and count of it
print("Unique Passenger Classes:", titanic["Pclass"].unique());
print("\nAge Value Counts:\n", titanic["PassengerAge"].value_counts().head());
print("\nPclass Value Counts:\n", titanic["Pclass"].value_counts());
```

```python
import pandas as pd

# Employee Data
EMP = pd.DataFrame({
    "EID": [1, 2, 3, 4],
    "Name": ["John", "Alice", "Sam", "Ravi"],
    "CID": [101, 102, 103, 104]   # Company IDs
})

# Company Data
COMP = pd.DataFrame({
    "CID": [101, 102, 104, 105],
    "Company": ["TCS", "Infosys", "Wipro", "HCL"]
})

# Save as Excel files
EMP.to_excel("EMP.xlsx", index=False);
COMP.to_excel("COMP.xlsx", index=False);

print("EMP.xlsx and COMP.xlsx created successfully!");
```

```python
# Load Excel files
emp_data = pd.read_excel("EMP.xlsx");
comp_data = pd.read_excel("COMP.xlsx");

# Inner Join
inner = pd.merge(emp_data, comp_data, on="CID", how="inner");
print("\nInner Join:\n", inner);

# Left Join
left = pd.merge(emp_data, comp_data, on="CID", how="left");
print("\nLeft Join:\n", left);

# Outer Join
outer = pd.merge(emp_data, comp_data, on="CID", how="outer");
print("\nOuter Join:\n", outer);
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cgpa1=[9.2,6.2,8.6,8.2,6.7,5.4,6,8];
cgpa2=[7,7.7,6.7,9,8,5.6,8,6];
cgpa3=[7.2,6.7,7.7,7,6,6.6,7,5.4];
cgpa4=[7.6,8.8,8.7,9,8,7.6,9,9.4];
sem=[1,2,3,4,5,6,7,8];
```

```python
plt.plot(sem, cgpa1);
plt.scatter(sem, cgpa2);

plt.show();
```

```python
plt.plot(sem, cgpa1);
plt.scatter(sem, cgpa2);
plt.ylim(0, 10);

plt.show();
```

```python
plt.title("Semester VS CGPA");
plt.xlabel("Semester");
plt.ylabel("CGPA");

plt.plot(sem, cgpa1);
plt.scatter(sem, cgpa2);

plt.ylim(0, 10);

plt.show();
```

```python
plt.xlabel('Semester');
plt.ylabel('CGPA');
plt.title('Semester VS CGPA');

x = np.arange(len(sem));
bar_width = 0.1;

plt.bar(x, cgpa1, width=bar_width, label="CGPA-1")
plt.bar(x + bar_width, cgpa2, width=bar_width, label="CGPA-2")
plt.bar(x + 2*bar_width, cgpa3, width=bar_width, label="CGPA-3")

plt.xticks(x + 1.5*bar_width, sem) ;

plt.ylim(0, 10);

plt.show();
```

```python
plt.plot(sem, cgpa1, label='L1')
plt.plot(sem, cgpa2, label='L2')
plt.plot(sem, cgpa3, label='L3')

# Adding labels and title
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.title('Semester VS CGPA')

plt.ylim(0, 10)


plt.legend()

plt.show()
```
